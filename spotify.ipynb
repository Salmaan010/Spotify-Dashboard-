Top 5 artist by popularity

```python
top_five_artist = df.groupby("artist_name").count().sort_values(by = "artist_popularity",ascending = False )["artist_popularity"][:5]
top_five_artist
```

```python
top_five_artist.plot.barh()
plt.show()
```

Top 5 most loudest tracks

```python

top_five_loudest_tracks = df[["loudness","track_name"]].sort_values(by = "loudness",ascending = True)[:5]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = "loudness" , y = "track_name",data = top_five_loudest_tracks)
plt.title("top 5 loudest tracks")
plt.show()

```

Artist with more dancebility tracks

```python
top_five_artist_danceable_tracks = df[["danceability","track_name","artist_name"]].sort_values(by = "danceability",ascending = False)[:5]
top_five_artist_danceable_tracks 
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = "danceability",y = "track_name",data = top_five_artist_danceable_tracks)
plt.title("Top 5 most danceable tracks")
plt.show()
```

Top  5 instrumentalness album and track name

```python

top_instrumental_album = df[["instrumentalness","track_name","album"]].sort_values(by= "instrumentalness",ascending = False)[:5]
top_instrumental_album
```

```python
plt.Figure(figsize=(12,7))
plt.pie(x = "instrumentalness",data = top_instrumental_album,autopct='%1.2f%%',labels=top_instrumental_album.album)
plt.title("TOP 5 instrumentl album")
plt.show()
```
